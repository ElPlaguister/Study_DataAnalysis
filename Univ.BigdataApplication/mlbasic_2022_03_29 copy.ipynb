```python
import numpy as np
import pandas as pd
```

```python
nbastat = pd.read_csv('nbastat2022.csv')
nbastat
```

```python
m = len(nbastat)
X = nbastat[['FGM', '3PA', 'FTA']].fillna(0)
Y = nbastat['PTS'].fillna(0)
```

```python
X = (np.array(X)).reshape(m, 3)
Y = (np.array(Y)).reshape(m, 1)
```

```python
X0 = np.ones((m, 1))
Xb = np.c_[X0, X]
```

```python
learning_rate = 0.0001
n_iter = 200

theta = np.zeros((4, 1))
gradients = np.zeros((4, 1))

for i in range(n_iter):
    gradients = (1.0/m)*Xb.T.dot(Xb.dot(theta) - Y)
    theta = theta - learning_rate * gradients
```

```python
X0 = np.ones((m, 1))
Xb = np.c_[X0, X]
Xb.shape
```

```python
import matplotlib.pyplot as plt

learning_rate = 0.0000001
n_iter = 4

theta = np.zeros((4, 1))
gradients = np.zeros((4, 1))

thetas = []
deltas = []
Js = []
for i in range(n_iter):
    cost = Xb.dot(theta) - Y
    gradients = (1.0/m)*Xb.T.dot(cost)
    theta = theta - learning_rate * gradients
    thetas.append(theta)
    #print(f'epoch: {i} ------------- \ndelta: {learning_rate * gradients}\n theta: {theta}')
    deltas.append(learning_rate * gradients)
    Js.append(sum(cost)[0])

new_n_iter = n_iter
thetas = np.array(thetas)
for i in range(4):
    plt.subplot(2, 2, i + 1)
    plt.plot(np.array(range(n_iter)), np.array(deltas)[:,i])
print(theta)
```
